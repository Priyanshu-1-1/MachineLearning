```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

## Pie Chart

```python
x1= [500,200,60,600,105]
my_labels=["A","B","C","D","E"]
```

```python
plt.pie(x1,labels=my_labels,autopct="%.2f%%")
plt.title("Pie Charts")
plt.show()
```
