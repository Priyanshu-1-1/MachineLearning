```python
import seaborn as sns 
```

```python
data= sns.load_dataset('tips')
```

```python
data.head()
```

```python
sns.relplot(x="total_bill",y="tip",data=data)
```

```python
sns.relplot(x="total_bill",y="tip",hue="sex",data=data)
```

```python
sns.relplot(x="total_bill",y="tip",hue="sex",size="size",data=data)
```

```python
sns.relplot(x="total_bill",y="tip",hue="sex",size="size",style="smoker",data=data)
```

```python
sns.catplot(x="day",y="tip")
```
